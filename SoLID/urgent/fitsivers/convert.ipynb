```python
import numpy as np
import pandas as pd
```

```python
table = pd.read_excel('2045.xlsx')
d1 = {}
d1['Q2'] = table['Q2'].values
d1['x'] = table['x'].values
d1['y'] = table['y'].values
d1['z'] = table['z'].values
d1['pT'] = table['pT'].values
d1['obs'] = table['obs'].values
d1['target'] = table['target'].values
d1['hadron'] = table['hadron'].values
d1['value'] = table['value'].values
d1['error'] = ((table['stat_u']**2 + table['systabs_u']**2)**0.5).values
#d1['error'] = table['stat_u']
d2 = pd.DataFrame(data=d1)
d2 = d2[['Q2','x','y','z','pT','obs','target','hadron','value','error']]
d2.to_csv('../expdata/sivers19.dat', sep='\t', index=False)
```
