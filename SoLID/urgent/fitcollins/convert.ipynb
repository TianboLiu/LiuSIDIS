```python
# solid
def convertsolid(infile, outfile):
    table = pd.read_csv(infile)
    d1 = {}
    d1['Q2'] = table['Q2'].values
    d1['x'] = table['x'].values
    d1['y'] = table['y'].values
    d1['z'] = table['z'].values
    d1['pT'] = table['pT'].values
    d1['obs'] = table['obs'].values
    d1['target'] = table['target'].values
    d1['hadron'] = table['hadron'].values
    d1['value'] = table['value'].values
    d1['stat'] = table['stat'].values
    d1['systrel'] = table['systrel'].values
    d1['systabs'] = table['systabs'].values
    d2 = pd.DataFrame(data=d1)
    d2 = d2[['Q2','x','y','z','pT','obs','target','hadron','value','stat','systrel','systabs']]
    d2.to_csv(outfile, sep='\t', index=False)
```

```python
convertsolid('../data/base0-neutron-pip.dat', '../expdata/base0-neutron-pip.dat')
convertsolid('../data/base0-neutron-pim.dat', '../expdata/base0-neutron-pim.dat')
convertsolid('../data/base0-proton-pip.dat', '../expdata/base0-proton-pip.dat')
convertsolid('../data/base0-proton-pim.dat', '../expdata/base0-proton-pim.dat')

convertsolid('../data/base1-neutron-pip.dat', '../expdata/base1-neutron-pip.dat')
convertsolid('../data/base1-neutron-pim.dat', '../expdata/base1-neutron-pim.dat')
convertsolid('../data/base1-proton-pip.dat', '../expdata/base1-proton-pip.dat')
convertsolid('../data/base1-proton-pim.dat', '../expdata/base1-proton-pim.dat')

convertsolid('../data/base2-neutron-pip.dat', '../expdata/base2-neutron-pip.dat')
convertsolid('../data/base2-neutron-pim.dat', '../expdata/base2-neutron-pim.dat')
convertsolid('../data/base2-proton-pip.dat', '../expdata/base2-proton-pip.dat')
convertsolid('../data/base2-proton-pim.dat', '../expdata/base2-proton-pim.dat')

convertsolid('../data/base3-neutron-pip.dat', '../expdata/base3-neutron-pip.dat')
convertsolid('../data/base3-neutron-pim.dat', '../expdata/base3-neutron-pim.dat')
convertsolid('../data/base3-proton-pip.dat', '../expdata/base3-proton-pip.dat')
convertsolid('../data/base3-proton-pim.dat', '../expdata/base3-proton-pim.dat')
```
