```python
import sys, os
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text',usetex=True)
```

# Load data

```python
xh = {}

xh['world'] = np.loadtxt('h1-world.dat', skiprows = 1)
xh['solid'] = np.loadtxt('h1-solid.dat', skiprows = 1)
xh['base'] = np.loadtxt('h1-base.dat', skiprows = 1)
```

```python
nrows, ncols = 2, 1
tol = 7.04
fig = plt.figure(figsize = (ncols * 7, nrows * 6))

ax = fig.add_subplot(nrows, ncols, 1)

ax.axhline(y = 0, linestyle = '-', color = 'black', alpha = 0.7)
ax.fill_between(xh['world'][:,0], xh['world'][:,1] - xh['world'][:,2] * tol, xh['world'][:,1] + xh['world'][:,2] * tol,\
               color = 'red', alpha = 0.5)
ax.fill_between(xh['world'][:,0], xh['world'][:,3] - xh['world'][:,4] * tol, xh['world'][:,3] + xh['world'][:,4] * tol,\
               color = 'blue', alpha = 0.5)
ax.fill_between(xh['solid'][:,0], xh['solid'][:,1] - xh['solid'][:,2] * tol, xh['solid'][:,1] + xh['solid'][:,2] * tol,\
               color = 'black', alpha = 0.5)
ax.fill_between(xh['solid'][:,0], xh['solid'][:,3] - xh['solid'][:,4] * tol, xh['solid'][:,3] + xh['solid'][:,4] * tol,\
               color = 'black', alpha = 0.5)
ax.plot(xh['world'][:,0], xh['world'][:,1], color = 'red', linestyle = '-')
ax.plot(xh['world'][:,0], xh['world'][:,3], color = 'blue', linestyle = '-')
ax.set_xlim(0,1)
ax.set_ylim(-0.4, 0.4)
ax.set_ylabel(r'$xh_1(x)$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.set_xticklabels([])
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.text(0.5, 0.95, r'$\rm World~vs.~SoLID(full)$', transform = ax.transAxes, horizontalalignment = 'center', fontsize = 25)

ax = fig.add_subplot(nrows, ncols, 2)
ax.plot(xh['world'][:,0], xh['world'][:,2] / xh['solid'][:,2],\
        linestyle = '-', linewidth = 5, color = 'red', alpha = 0.5, label = r'$u$')
ax.plot(xh['world'][:,0], xh['world'][:,4] / xh['solid'][:,4],\
        linestyle = '--', linewidth =5, color = 'blue', alpha = 0.5, label = r'$d$')
ax.set_xlim(0,1)
ax.set_ylim(0, 50)
ax.set_xlabel(r'$x$', size = 30)
ax.set_ylabel(r'$\rm Error^{World}/Error^{SoLID}$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.axhline(y = 10, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 20, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 30, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 40, linestyle = ':', color = 'black', alpha = 0.5)
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.legend(frameon = False, loc = 'upper right', fontsize = 30)

fig.tight_layout()
fig.savefig("world-full.pdf")
```

```python
nrows, ncols = 2, 1
tol = 7.04
fac = 1.022
fig = plt.figure(figsize = (ncols * 7, nrows * 6))

ax = fig.add_subplot(nrows, ncols, 1)

ax.axhline(y = 0, linestyle = '-', color = 'black', alpha = 0.7)
ax.fill_between(xh['world'][:,0], xh['world'][:,1] - xh['world'][:,2] * tol, xh['world'][:,1] + xh['world'][:,2] * tol,\
               color = 'red', alpha = 0.5)
ax.fill_between(xh['world'][:,0], xh['world'][:,3] - xh['world'][:,4] * tol, xh['world'][:,3] + xh['world'][:,4] * tol,\
               color = 'blue', alpha = 0.5)
ax.fill_between(xh['base'][:,0], xh['base'][:,1] - xh['base'][:,2] * tol * fac, xh['base'][:,1] + xh['base'][:,2] * tol * fac,\
               color = 'black', alpha = 0.5)
ax.fill_between(xh['base'][:,0], xh['base'][:,3] - xh['base'][:,4] * tol * fac, xh['base'][:,3] + xh['base'][:,4] * tol * fac,\
               color = 'black', alpha = 0.5)
ax.plot(xh['world'][:,0], xh['world'][:,1], color = 'red', linestyle = '-')
ax.plot(xh['world'][:,0], xh['world'][:,3], color = 'blue', linestyle = '-')
ax.set_xlim(0,1)
ax.set_ylim(-0.4, 0.4)
ax.set_ylabel(r'$xh_1(x)$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.set_xticklabels([])
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.text(0.5, 0.95, r'$\rm SoLID(Baseline)~vs.~World$', transform = ax.transAxes, horizontalalignment = 'center', fontsize = 25)



ax = fig.add_subplot(nrows, ncols, 2)
ax.plot(xh['world'][:,0], xh['world'][:,2] / xh['base'][:,2] / fac,\
        linestyle = '-', linewidth = 5, color = 'red', alpha = 0.5, label = r'$u$')
ax.plot(xh['world'][:,0], xh['world'][:,4] / xh['base'][:,4] / fac,\
        linestyle = '--', linewidth =5, color = 'blue', alpha = 0.5, label = r'$d$')
ax.set_xlim(0,1)
ax.set_ylim(0, 50)
ax.set_xlabel(r'$x$', size = 30)
ax.set_ylabel(r'$\rm Error^{World}/Error^{Baseline}$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.axhline(y = 10, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 20, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 30, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 40, linestyle = ':', color = 'black', alpha = 0.5)
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.legend(frameon = False, loc = 'upper right', fontsize = 30)


fig.tight_layout()
fig.savefig("trans-world-base.pdf")
```

```python
nrows, ncols = 2, 1
tol = 7.04
fac = 1.022
fig = plt.figure(figsize = (ncols * 7, nrows * 6))

ax = fig.add_subplot(nrows, ncols, 1)

ax.axhline(y = 0, linestyle = '-', color = 'black', alpha = 0.7)
ax.fill_between(xh['base'][:,0], xh['base'][:,1] - xh['base'][:,2] * tol * fac, xh['base'][:,1] + xh['base'][:,2] * tol * fac,\
               color = 'red', alpha = 0.5)
ax.fill_between(xh['base'][:,0], xh['base'][:,3] - xh['base'][:,4] * tol * fac, xh['base'][:,3] + xh['base'][:,4] * tol * fac,\
               color = 'blue', alpha = 0.5)
ax.fill_between(xh['solid'][:,0], xh['solid'][:,1] - xh['solid'][:,2] * tol, xh['solid'][:,1] + xh['solid'][:,2] * tol,\
               color = 'black', alpha = 0.5)
ax.fill_between(xh['solid'][:,0], xh['solid'][:,3] - xh['solid'][:,4] * tol, xh['solid'][:,3] + xh['solid'][:,4] * tol,\
               color = 'black', alpha = 0.5)
ax.plot(xh['base'][:,0], xh['base'][:,1], color = 'red', linestyle = '-')
ax.plot(xh['base'][:,0], xh['base'][:,3], color = 'blue', linestyle = '-')
ax.set_xlim(0,1)
ax.set_ylim(-0.4, 0.4)
ax.set_ylabel(r'$xh_1(x)$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.set_xticklabels([])
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.text(0.5, 0.95, r'$\rm Enhanced~vs.~Baseline$', transform = ax.transAxes, horizontalalignment = 'center', fontsize = 25)



ax = fig.add_subplot(nrows, ncols, 2)
ax.plot(xh['base'][:,0], xh['base'][:,2] * fac / xh['solid'][:,2],\
        linestyle = '-', linewidth = 5, color = 'red', alpha = 0.5, label = r'$u$')
ax.plot(xh['base'][:,0], xh['base'][:,4] * fac / xh['solid'][:,4],\
        linestyle = '--', linewidth =5, color = 'blue', alpha = 0.5, label = r'$d$')
ax.set_xlim(0,1)
ax.set_ylim(0, 5)
ax.set_xlabel(r'$x$', size = 30)
ax.set_ylabel(r'$\rm Error^{Baseline}/Error^{Enhanced}$', size = 30)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', length = 5, pad = 6, labelsize = 25)
ax.axhline(y = 1, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 2, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 3, linestyle = ':', color = 'black', alpha = 0.5)
ax.axhline(y = 4, linestyle = ':', color = 'black', alpha = 0.5)
ax.axvline(x = 0.05, linestyle = '--', color = 'black', alpha = 0.7)
ax.axvline(x = 0.6, linestyle = '--', color = 'black', alpha = 0.7)
ax.legend(frameon = False, loc = 'upper right', fontsize = 30)


fig.tight_layout()
fig.savefig("trans-base-enhanced.pdf")
```
